```python
import csv
import pandas as pd

path='aaa.csv'
with open(path) as f:
    reader = csv.reader(f, delimiter="\t")
    d = list(reader)
```

```python
new=d[1:]
keys=d[0]
```

```python
apidict = {x:list(y) for x,y in zip(keys, zip(*new))}
```

```python
df=pd.DataFrame(apidict)
```

```python
del df['Animal_ID']
```

```python
del df['Month_Day_Year']
```
